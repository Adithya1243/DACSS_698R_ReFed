```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.patches as mpatches
```

```python
# Load data into a DataFrame
data = pd.read_csv('ReFED_US_Food_Surplus_Summary_Umass_2019.csv')
# Display first few rows of data
print(data.head())
```

```python
# Check for missing values
print(data.isnull().sum())
```

```python
# Explore distribution of types of donated food
plt.figure(figsize=(10,6))
sns.countplot(x='food_type', data=data)
plt.title('Distribution of Donated Food Types')
plt.xlabel('Food Type')
plt.ylabel('Count')
plt.show()
```

```python
# Assuming 'data' is a pandas DataFrame
n = data.groupby('food_type')['tons_supply'].sum()

# Create a bar chart
fig, ax = plt.subplots()
n.plot(kind='bar', ax=ax)

# Set chart title and labels
ax.set_title('Total Tons Supply per Food Type')
ax.set_xlabel('Food Type')
ax.set_ylabel('Tons Supply')

# Show the chart
plt.show()
```

```python
# Assuming 'data' is a pandas DataFrame
n = data.groupby('food_type')['tons_surplus'].sum()

# Create a bar chart
fig, ax = plt.subplots()
n.plot(kind='bar', ax=ax)

# Set chart title and labels
ax.set_title('Total Tons Surplus per Food Type')
ax.set_xlabel('Food Type')
ax.set_ylabel('Tons Surplus')

# Show the chart
plt.show()
```

```python

# Assuming 'data' is a pandas DataFrame
n = data.groupby('sector')['tons_supply'].sum()

# Create a bar chart
fig, ax = plt.subplots()
n.plot(kind='bar', ax=ax)

# Set chart title and labels
ax.set_title('Total Tons Supply per sector')
ax.set_xlabel('sector')
ax.set_ylabel('Tons Supply')

# Show the chart
plt.show()
```

```python
# Assuming 'data' is a pandas DataFrame
n = data.groupby('sector')['tons_surplus'].sum()

# Create a bar chart
fig, ax = plt.subplots()
n.plot(kind='bar', ax=ax)

# Set chart title and labels
ax.set_title('Total Tons Surplus per sector')
ax.set_xlabel('sector')
ax.set_ylabel('Tons Surplus')

# Show the chart
plt.show()
```

```python
# Calculate summary statistics for donation amount by donor type
print(data.groupby('sector')['tons_supply'].describe())
```

```python
# Calculate summary statistics for donation amount by donor type
print(data.groupby('sector')['tons_surplus'].describe())
```

```python
# Calculate summary statistics for donation amount by food type
print(data.groupby('food_type')['tons_supply'].describe())
```

```python
# Calculate summary statistics for donation amount by food type
print(data.groupby('food_type')['tons_surplus'].describe())
```

```python

# set plot style: grey grid in the background:
sns.set(style="darkgrid")

# set the figure size
plt.figure(figsize=(14, 14))

# Assuming 'data' is a pandas DataFrame
x = data.groupby('food_type')['tons_supply'].sum()

# Create a bar chart
fig, ax = plt.subplots()
bar1 = x.plot(kind='bar', ax=ax, color = 'lightgreen')

# Assuming 'data' is a pandas DataFrame
y = data.groupby('food_type')['tons_surplus'].sum()

# Create a bar chart
fig, ax = plt.subplots()
bar2 = y.plot(kind='bar', ax=ax, color = 'red')

# add legend
top_bar = mpatches.Patch(color='lightgreen', label='tons supply')
bottom_bar = mpatches.Patch(color='red', label='tons surplus')
plt.legend(handles=[top_bar, bottom_bar])

# show the graph
plt.show()
```
